```python
import numpy as np
from scipy.special import comb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
%%time
S=10000
full=[[0]]
for s in range(1,S):
    layer=[0]
    for t in range (1,s):
        u=full[s-1][t]
        ur=full[s-1][t-1]
        layer.append(u*(2*s-2-t)/(2*s-2)+ur*(t-1)/(2*s-2))
    layer.append(1/2**(s-1))
    full.append(layer)
#print(full)
```

```python
%%time
exp=[]
for f in full:
    exp.append(np.dot(f,np.arange(len(f))))
exp=np.array(exp)
```

```python
plt.figure(figsize=(16,8))
res=exp**2
_=plt.plot(res)
```
