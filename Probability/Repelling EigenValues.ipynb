### Repelling EigenValues

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N=1000
```

```python
A=np.random.normal(0,1,(N,N))
spec_A=np.linalg.eigvals(A)
```

```python
real=spec_A.real
imag=spec_A.imag
```

```python
B=2*np.random.rand(2*N,2)-1
B=B[B[:,0]**2+B[:,1]**2<1][:N]
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.scatter(real,imag)
plt.title('{} Complex EigenValues!'.format(N))
plt.subplot(1,2,2)
plt.scatter(B[:,0],B[:,1])
plt.title('{} Uniform Random Points in a Circle'.format(N))
```

It is counter intutive that eigenvalues are more well distributed than the uniform distribution!
