```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N=10000000
max_dim=15

points=np.random.rand(N,max_dim)
x=range(1,max_dim+1)
```

```python
%%time
vols=[]
for d in x:
    mask= np.sum(points[:,:d]**2,1)<1
    vols.append(2**d*np.sum(mask)/N)
```

```python
plt.figure(figsize=(15,6))
_=plt.scatter(x,vols)
```
