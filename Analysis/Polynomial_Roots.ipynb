# Polynomial Roots
-----

I learned this amazing phenomena from John Baez [homepage](http://math.ucr.edu/home/baez/roots/). Consider all polynomials of deg $\leqslant N$ (where N is by default 12) whose coefficients are $\pm \frac12$ and draw all the roots in the imaginary plane.

That's it.

You will see the miracle.

You may see some stunning pictures in Baez homepage and of course read some explanations. But I recommend to play around with the parameters below to create your own plot.

For a high quality version of the picture see [this](http://math.ucr.edu/home/baez/roots/polynomial_roots.png).

|Parameter|Explanation|
|:------:|:----:|
|N| max degree of polynomials |
|b| base , the coefficents of polynomials will be of a symmetric arithmetic progression around zero of length b |
|p|I will use $L_p$ distance from origin to color the points |
|num_of_plots| Number of plots|

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N=12
b=2
p=0.5
num_of_plots=1
cmaps=['magma','hot','CMRmap', 'Paired',  'Spectral',  'afmhot',  'copper', 'inferno', 'viridis']
```

```python
xroots=[]
yroots=[]
for i in range(b**(N+1)):
    rpr=np.base_repr(i,b)
    coef=[int(rpr[j])-(b-1)/2 for j in range(len(rpr))]
    roots=np.roots(coef)
    for k in roots:
        xroots.append(np.real(k))
        yroots.append(np.imag(k))
        
c=(np.abs(np.array(yroots))**p+np.abs(np.array(xroots))**p)**(1/p)
```

```python
for cmap in cmaps[:num_of_plots]:
    plt.figure(figsize=(15,10))
    plt.axis('off')
    plt.title(cmap)
    _=plt.scatter(xroots,yroots,s=0.2,c=c,cmap=cmap)
```
