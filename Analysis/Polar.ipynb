# Animated Funtions
------------

This notebook is devoted to animated plots. My main focus are on trigonometric functions which are inspired by polar coordinates. You may tune the function randomly and get surprise by the shape of the function.

I've made up some interesting plots. I recommend to try it yourself. You need to download the notebook to make changes.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib notebook

th=np.arange(0,np.pi,0.01)
```

Here is the definition of the function. $\theta$ changes in the interval $(0,\pi)$ and the range of $t$, the time, can be tuned in the plotting cell.

```python
def func(t):
    x=np.cos(t*th)*np.cos(th)
    y=np.sin(t*th)*np.sin(th)
    return x,y
```

```python
def animate(c):
    x,y=func(c)
    ax.set_xlim(-1.1,1.1)
    ax.set_ylim(-1.1,1.1)
    line.set_xdata(x)
    line.set_ydata(y)
```

```python
fig,ax=plt.subplots()
plt.title('($\cos(t a )\cos(a),\cos(ta)\sin(a)$)')
x,y=func(1)
line,=ax.plot(x,y)
ax.axis('off')



ani=FuncAnimation(fig,animate,np.arange(1,20,0.05),interval=2)
```
